```python
#List: is a collection which is ordered and changeable(modifiable). Allows duplicate members.
# possible to append in the existing list
fruits = ['banana', 'orange', 'mango', 'lemon','apple']                     # list of fruits
countries = ['India', 'USA', 'UK', 'Sweden', 'Qatar']

# Print the lists and its length
print('Fruits:', fruits)
print('Number of fruits:', len(fruits))

print('Countries:', countries)
print('Number of countries:', len(countries))

#accessing the element based on the index position.
fruits = ['banana', 'orange', 'mango', 'lemon','apple']
first_fruit = fruits[-4]
last_fruit = fruits[-1]
second_last = fruits[-2]
print(first_fruit)      # banana
print(last_fruit)       # apple
print(second_last)      # lemon

#find the element is exists or not
fruits = ['banana', 'orange', 'mango', 'lemon','apple']
does_exist = 'banana' in fruits
print(does_exist)  # True
does_exist = 'lime' in fruits
print(does_exist)  # False
```

```python
#Adding the elements in the list
#append method - add the element

fruits = ['banana', 'orange', 'mango', 'lemon']
fruits.append('pineapple')
print(fruits)           
fruits.append('grapes')  
print(fruits)
```

```python
#adding and Remove element in the list by using append method
n=int(input())
a=list() #create the empty list
for i in range(0,n):
    app_str='Added element in Postion-'+str(i)
    a.append(app_str)
print('Print the entire list after added all the Elements....')
print (a)

#remove the element based on the element value
print('Removed the Element by Element...')
for i in range(0,n):
    app_str='Added element in Postion-'+str(i)
    a.remove(app_str)
    print(a)
    
#REMOVEED THE ELEMENT BASED ON THE INDEX...

for i in range(0,n):
    app_str='Added element in Postion-'+str(i)
    a.append(app_str)
print('Print the entire list after added all the Elements....')
print (a)

a.pop() # removed the last element.. 
print('Removed the last Element....')
print(a)
a.pop(1) #removed the element based on the index...
print('Removed the element at index position is 1')
print(a)

#Counting Items in a List by using the count method. 
print('count of the list...')
print(a.count('Added element in Postion-2'))


#Removing Items Using Del...
#the del keyword removes the specified index and it can also be used to delete items within index range. It can also delete the list completely....

for i in range(0,n):
    app_str='Added element in Postion-'+str(i)
    a.append(app_str)
print('Print the entire list after added all the Elements....')
print (a)
del a[0]
print('Removed the first eleemnt based on the index position...')
print (a)
del a[1:3] # remove the element based on the index range...
print('Removed the eleemnts based index range...')
print (a)
del a
#print(a) # it will throw the error name 'a' is not defined. Since, list a is removed.

#Reversing a List
a=['India','Srilanka','USA','UK','Qatar']
print(a)
a.reverse()
print('After done the reverse method...')
print(a)

#soring the element by using the sort method.

a.sort() #ascending order...
print('after done the sorting(ascending)...')
print(a)
a.sort(reverse=True)
print('after done the sorting(Desending order)...')
print(a)

#joining two lists...
#adding elements in the X and Y... and wil join these 2 lists into Z list.
x=list()
y=list()
for i in range(0,n):
    x.append(i)
print('Print the X List...')
print(x)
i=len(x)
while(i>0):
    y.append(i)
    i=i-1
print('Print the Y list....')
print(y)
print('Join these 2 List and store into Z...')
z=x+y
print(z)



```
